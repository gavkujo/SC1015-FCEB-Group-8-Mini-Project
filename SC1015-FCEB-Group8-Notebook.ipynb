## Linear Regression

As seen above, we see that the variables "P_SEMI_MAJOR_AXIS", "P_ECCENTRICITY", "S_MAG", "S_METALLICITY", "S_TYPE" and "S_AGE" have NULL values in their entries, so we will not be using these variables in our regression model.

Since Linear Regression is a Numeric In - Numeric Out Model, we will be using the 16 Numeric Variables to run this model.


```python
# Import essential models and functions from sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Extract Response and Predictors
y = pd.DataFrame(MainData["P_ESI"])
X = pd.DataFrame(MainData[["S_TIDAL_LOCK", "S_ABIO_ZONE", "S_SNOW_LINE", "S_LUMINOSITY", "P_TEMP_EQUIL", "P_FLUX", "P_DISTANCE", "P_DENSITY", "P_GRAVITY", "P_POTENTIAL", "S_TEMPERATURE", "S_RADIUS", "S_MASS", "P_RADIUS", "P_MASS", "P_PERIOD"]])

# Split the Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

# Linear Regression using Train Data
linreg = LinearRegression()         # create the linear regression object
linreg.fit(X_train, y_train)        # train the linear regression model

# Coefficients of the Linear Regression line
print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
print()

# Print the Coefficients against Predictors
print(pd.DataFrame(list(zip(X_train.columns, linreg.coef_[0])), columns = ["Predictors", "Coefficients"]))
print()

# Predict Response corresponding to Predictors
y_train_pred = linreg.predict(X_train)
y_test_pred = linreg.predict(X_test)

# Plot the Predictions vs the True values
f, axes = plt.subplots(1, 2, figsize=(24, 12))
axes[0].scatter(y_train, y_train_pred, color = "blue")
axes[0].plot(y_train, y_train, 'w-', linewidth = 1)
axes[0].set_xlabel("True values of the Response Variable (Train)")
axes[0].set_ylabel("Predicted values of the Response Variable (Train)")
axes[1].scatter(y_test, y_test_pred, color = "green")
axes[1].plot(y_test, y_test, 'w-', linewidth = 1)
axes[1].set_xlabel("True values of the Response Variable (Test)")
axes[1].set_ylabel("Predicted values of the Response Variable (Test)")
plt.show()

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(X_train, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_pred))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", linreg.score(X_test, y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_pred))
print()
```

As we can see, we ended up with a really low MSE (~0.005) and a high Explained Variance (~0.7) suggesting that this model itself is pretty good for predicting the ESI for the planets given these variables.

However, As we saw when we were Exploring the Data, P_TYPE has a significant correlation with P_ESI and hence should be included in the Prediction process. However, P_TYPE is a categorical variable and Linear Regression is a Numeric in - Numeric out model.

Hence we shall use One Hot Encoding Technique to encode the categorical variable in a way such that the Linear Regression model can interpret the Categorical Information.

## One Hot Encoder

```python
# One Hot Encoder Linear Regression for the Categorical Variable P_TYPE

from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error

y = pd.DataFrame(MainData["P_ESI"])
X = pd.DataFrame(MainData[["S_TIDAL_LOCK", "S_ABIO_ZONE", "P_DISTANCE", "P_DENSITY", "P_GRAVITY", "S_TEMPERATURE", "P_RADIUS", "P_MASS", "P_PERIOD", "P_TYPE"]])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

encoder_one_hot = OneHotEncoder()
X_train_one_hot = encoder_one_hot.fit_transform(X_train[['P_TYPE']])

model_one_hot = LinearRegression().fit(X_train_one_hot, y_train)

X_test_one_hot = encoder_one_hot.transform(X_test[['P_TYPE']])
y_pred_one_hot = model_one_hot.predict(X_test_one_hot)
mse_one_hot = mean_squared_error(y_test, y_pred_one_hot)

f, axes = plt.subplots(1, 2, figsize=(24, 12))
axes[0].scatter(y_train, y_train_pred, color = "blue")
axes[0].plot(y_train, y_train, 'w-', linewidth = 1)
axes[0].set_xlabel("True values of the Response Variable (Train)")
axes[0].set_ylabel("Predicted values of the Response Variable (Train)")
axes[1].scatter(y_test, y_test_pred, color = "green")
axes[1].plot(y_test, y_test, 'w-', linewidth = 1)
axes[1].set_xlabel("True values of the Response Variable (Test)")
axes[1].set_ylabel("Predicted values of the Response Variable (Test)")
plt.show()

'''
# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", LinearRegression().score(X_train_one_hot, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_pred))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", LinearRegression().score(X_test_one_hot, y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_pred))
print()
'''

print(f"One-Hot Encoding Model - Mean Squared Error: {mse_one_hot}")
```

What we did above was convert a dataframe with just the categorical variable P_TYPE data and encode that into a matrix-type data: X_Train_one_hot and X_test_one_hot.
We then used this data and fed it into the Linear Regression Model to predict the ESI values.

The low MSE values (~0.003) implies that this encoding technique is efficient.

## One Hot Encoding Linear Regression

We will now proceed to encode the entire usable dataframe (the 16 Numerical variables and P_TYPE) to predict the ESI values using Linear Regression.

```python
# Standard Linear Regression

MainData['P_TYPE'].unique()
y = pd.DataFrame(MainData["P_ESI"])
X = pd.DataFrame(MainData[["S_TIDAL_LOCK", "S_ABIO_ZONE", "P_DISTANCE", "P_DENSITY", "P_GRAVITY", "S_TEMPERATURE", "P_RADIUS", "P_MASS", "P_PERIOD", "P_TYPE"]])
encoder_one_hot = OneHotEncoder()
X = encoder_one_hot.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

print(regressor.coef_)
print(regressor.intercept_)

y_pred = regressor.predict(X_test)

f, axes = plt.subplots(1, 1, figsize=(24, 12))
axes.scatter(y_test, y_pred, color = "green")
axes.plot(y_test, y_test, 'w-', linewidth = 1)
axes.set_xlabel("True values of the Response Variable (Test)")
axes.set_ylabel("Predicted values of the Response Variable (Test)")
plt.show()

from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

print('Mean Absolute Error = ',mean_absolute_error(y_test,y_pred))
print('R2 score = ',r2_score(y_test,y_pred))
```
